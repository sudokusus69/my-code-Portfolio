```python
import pandas as pd 

ds = pd.read_csv('subs.csv')

x = ds.iloc[:,:-2]
y = ds['sal']

import matplotlib.pyplot as plt 

```

# GRID SEARCH CV


```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)


from sklearn.tree import DecisionTreeRegressor, plot_tree

# dt = DecisionTreeRegressor(criterion='squared_error', max_depth=15,splitter='best')
# dt.fit(x_train,y_train)
# print(dt.score(x_test,y_test)*100, dt.score(x_train,y_train)*100)


from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

dic = {'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
        'splitter': ['best', 'random'],
        'max_depth' : [i for i in range(2,20)]}

gd = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid=dic)
gd.fit(x_train,y_train)

print(gd.best_params_)
print(gd.best_score_*100)
```

# RANDOMISED SEARCH CV


```python
rd = RandomizedSearchCV(DecisionTreeRegressor(random_state=42), param_distributions=dic, n_iter=20)
rd.fit(x_train,y_train)
rd.score(x_test,y_test)*100

print(rd.best_params_)
print(rd.best_score_)
```
