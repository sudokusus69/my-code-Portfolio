# **OBJECT ORIENTED PROGRAMMING (OOPS)**

```python
class Student:
    
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        
    def welcome(self):
        print('welcome ', self.name)

    def get_marks(self):
        return self.marks

s1 = Student('yass', 67)

s1.welcome()

s1.get_marks()
```

```python
class St:

    
    def __init__(self, name, math, phy, chem):
        self.name = name
        self.math = math
        self.phy = phy
        self.chem = chem

    def avg(self):
        print((self.math + self.phy + self.chem) / 3 )
        

s1 = St('yass', 67, 68, 69)
s1.avg()



### Different Approach

class St:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def avg(self):
        sum = 0 
        for x in self.marks:
            sum += x 
        print(sum/3)
        

        
s1 = St('yass', [67, 68, 69])

s1.avg()
```

## **STATIC METHOD ...**

```python
# Using object that are not really work under object means
#  they do not take self parameters as usual but they work under class and not def ,
''' that is why they can be used without any self input and are called Static methods'''

'''for this we are going to simply use decorator which converts a function in to a static method'''

class St:

    @staticmethod 
    def hello():
        print('hello gng')


St.hello()
```
