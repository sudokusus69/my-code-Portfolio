# NAIVE BAYES
TRAINING DATAS WITH DIFFERENT DISTRIBUTION IN NAIVE BAYES

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions


ds = pd.read_csv('cgpa.csv')

ds.head(3)

x=ds.iloc[:,:-1]
y=ds['placed']

from sklearn.model_selection import train_test_split
x_train, x_test , y_train ,y_test = train_test_split(x,y, test_size=0.2, random_state=42)

from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB

# GUASSION DISTRIBUTION
gnb = GaussianNB()
gnb.fit(x_train,y_train)
print(gnb.score(x_test,y_test)*100)

# MULTINOMIAL DISTRIBUTION
mul = MultinomialNB()
mul.fit(x_train,y_train)
print(mul.score(x_test,y_test)*100)

# BERNOULLI DISTRIBUTION
ber = BernoulliNB()
ber.fit(x_train,y_train)
print(ber.score(x_test,y_test)*100)

plt.figure(figsize=(4,3))
sns.scatterplot(x=ds['cgpa'], y=ds['score'], hue=ds['placed'])
plt.show()
```

# PLOTTING ALL 3 DISTRIBUTIONS

```python
from mlxtend.plotting import plot_decision_regions

fig, axes = plt.subplots(1,3, figsize=(15,4))

# GUASSION DISTRIBUTION
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb, ax=axes[0])


# MULTINOMIAL DISTRIBUTION
x2 = plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mul, ax=axes[1])


# GUASSION DISTRIBUTION
x3 = plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=ber, ax = axes[2])

plt.show()
```
