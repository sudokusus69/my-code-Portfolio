# CLASSIFICATION DECISION TREE

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

ds = pd.read_csv('cgpa.csv')

x = ds.iloc[:,:-1]
y = ds['placed']


from sklearn.preprocessing import StandardScaler

st = StandardScaler()
st.fit(x)
x = pd.DataFrame(st.transform(x), columns=x.columns)
print(x.head(3))

from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini')
dt.fit(x_train,y_train)

print(dt.score(x_test,y_test)*100)
print(dt.score(x_train,y_train)*100) # the more the difference btw the train and test scores
# the more is the overfitting in the data
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(3,2))
plot_tree(dt)
plt.tight_layout()
plt.show()


from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(3,2))
plot_decision_regions(x.to_numpy(), y.to_numpy(),clf=dt)
plt.show()
```

```python
for i in range(1,13):

    dt1 = DecisionTreeClassifier(max_depth=i)
    dt1.fit(x_train,y_train)


    k2 = dt1.score(x_train,y_train)*100
    k1 = dt1.score(x_test,y_test)*100

    print( i, k2, k1)


dt2 = DecisionTreeClassifier(max_depth=3)
dt2.fit(x_train,y_train)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

plot_tree(dt2, ax=axes[0])
axes[0].set_title("Decision Tree (max_depth=3)")

plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=dt2, ax=axes[1])
axes[1].set_title("Decision Regions (max_depth=3)")
plt.tight_layout()

plt.show()
    
```

# REGRESSION DECISION TREE


```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

ds = pd.read_csv('cgpa.csv')

x = ds.iloc[:,:-1]
y = ds['placed']


from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)


from sklearn.tree import DecisionTreeClassifier, plot_tree

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

print(dt.score(x_test,y_test)*100)
print(dt.score(x_train,y_train)*100)

```

```python
plt.figure(figsize=(4,3))
plot_tree(dt)
plt.show()
```
