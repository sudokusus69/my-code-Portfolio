# CLUSTERING

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

ds = pd.read_csv('flower.csv')

sns.pairplot(ds, height=1)
plt.show()

from sklearn.cluster import KMeans
wcss = []

for i in range(2,20):
    km = KMeans(init='k-means++', n_clusters=i)
    km.fit(ds)
    wcss.append(km.inertia_)


plt.plot([i for i in range(2,20)], wcss)
plt.xticks([i for i in range(2,20)])
plt.scatter([i for i in range(2,20)], wcss)
plt.grid(axis='x')
plt.show()

# sns.pairplot(ds)
```

# DBSCAN 

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN


# it is used to create a dataset with two moon like shapes
x,y = make_moons(noise = 0.05, n_samples=250)

# it is used to create 3 column fields beased on the above dataset
# then it is converted to a pandas dataframe
ds = {'d1' : x[:,0], 'd2':x[:,1], 'output':y}
df = pd.DataFrame(ds)


# ti is used to train the DBSCAN model
db = DBSCAN(eps= 0.5, metric='euclidean', min_samples=5)
df['prd'] = db.fit_predict(df)


# it is used to plot the clustering data 
sns.scatterplot(x= 'd1', y='d2',data=df,  hue= 'prd')
```

# APRIORI and FP GROWTH ALGORITHM

```python

from mlxtend.frequent_patterns import apriori
import numpy as np
import collections
from mlxtend.preprocessing.transactionencoder import TransactionEncoder

'''FP-Growth (Frequent Pattern Growth) is an algorithm used to find groups of items that often appear together in large datasets,
like shopping baskets.
 It is similar to the Apriori algorithm but is much faster because it doesnt check every possible combination. 
Instead, it builds a special tree (called an FP-tree) to store item information and quickly find frequent patterns.

1) FREQUENT PATTERN SET 
2) ORDERED ITEM SET 
3) ORDERED ITEM SET AND CONDITIONAL FREQUENT PATTERN TREE IS BUILT
4) FREQUENT PATTERN RULES


Imagine you have a long list of shopping carts from a supermarket, and you want to know which products are frequently bought together.
The Apriori algorithm looks through all the carts and finds groups of items that often appear together.'''

ds = pd.read_csv('transactions.csv')

# Remove the 'TransactionID' column and its values from the dataframe
ds.drop('TransactionID', axis=1, inplace=True)



# This loop goes through each row in the dataset and collects all the items bought in that transaction.
cus = []
for i in range(0, ds.shape[0]):
    sus = []
    for j in ds.columns:
        if type(ds[j][i]) == str:
            sus.append(ds[j][i])
    cus.append(sus)


# This part counts how many times each item appears in all transactions.
l = []
for i in cus:
    for j in i:
        l.append(j)

p = collections.Counter(l)

# This creates a DataFrame showing each item and how many times it was bought.
d = {'item': p.keys(), 'values': p.values()}
pd.DataFrame(d).sort_values(by='values', ascending=False)

# TransactionEncoder changes the list of items into a table of True/False values (item bought or not).
tr = TransactionEncoder()
tr.fit(cus)
df = pd.DataFrame(tr.transform(cus), columns=tr.columns_)
print(df.head())

# The apriori function finds groups of items that are often bought together.
apriori(df, min_support=0.07, use_colnames=True, max_len=3).sort_values(by='support')
```

```python
# fp growth
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df, min_support=0.07, use_colnames=True, max_len= 3).sort_values(by = 'support')
```
