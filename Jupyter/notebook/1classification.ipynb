# LINEAR CLASSIFICATION



```python
ds = pd.read_csv('car_data2.csv')

x = ds.iloc[:,:-1]
y = ds['placed']


from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)


from sklearn.linear_model import LogisticRegression

lo = LogisticRegression()
lo.fit(x_train,y_train)
lo.score(x_test, y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
import numpy as np

plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lo)
plt.show()
```

```python
print(lo.coef_)
print(lo.intercept_)
```


# POLYNOMIAL CLASSIFICATION


```python
ds = pd.read_csv('data2.csv')

x = ds[['age_of_car', 'driven_kms']]
y = ds['bought']

from sklearn.preprocessing import PolynomialFeatures
po = PolynomialFeatures(degree=4)
po.fit(x)
x = pd.DataFrame(po.transform(x))


x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)


lo = LogisticRegression()
lo.fit(x_train, y_train)
print(lo.score(x_test,y_test)*100)

# sns.scatterplot(x=ds['age_of_car'], y=ds['driven_kms'], hue=ds['bought'])
```
