```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions

ds = pd.read_csv(r'D:\python\Machine_Learning\l\Jupyter notebook\subs.csv')
ds.shape

plt.figure(figsize=(4,3))
sns.scatterplot(x='age', y = 'sal', data=ds, hue='purchased')



x = ds.iloc[:,:-1]
y = ds.iloc[:,-1]

from sklearn.model_selection import train_test_split

x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)


from sklearn.linear_model import Perceptron

# Initialize Perceptron with some common parameters for better control
pt = Perceptron()
pt.fit(x_train, y_train)

print(pt.score(x_train,y_train)*100, round(pt.score(x_test,y_test)*100,2))


plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf = pt)
plt.show()
```
