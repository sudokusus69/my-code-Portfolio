# SIMPLE LINEAR REGRESSION


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

ds = pd.read_csv('car_dataset.csv')
x = ds[['present_price']]
y = ds['selling_price']

x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(x_train,y_train)



plt.scatter(x,y)
plt.plot(x, lr.predict(x))
plt.show()
```

```python
print(lr.score(x_test,y_test)*100)
print(lr.predict([[300]]))
```

# MULTIPLE LINEAR REGRESSION

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

ds = pd.read_csv('car_dataset.csv')

x = ds.iloc[:,2:4]
y = ds['selling_price']

x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(x_train,y_train)

```

```python
print(lr.score(x_test,y_test)*100)
```

# POLYNOMIAL REGRESSION


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


ds = pd.read_csv('car_data2.csv')

x = ds[['present_price']]
y = ds['selling_price']


# for polynomial regression training
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)

# calling train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

# for linear regression training
lr = LinearRegression()
lr.fit(x_train,y_train)

# final prediction for y axis 
prd = lr.predict(x)

plt.scatter(ds['present_price'], ds['selling_price'])
plt.plot(ds['present_price'], prd, c = 'red')
plt.show()


```

```python

test = pf.transform([[9.5]])
lr.predict(test)
```
