# CONFUSION MATRIX 
to analyse our data better using recall, f1 and precision scores

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

```

```python
ds = pd.read_csv('data2.csv')


ds.dropna(inplace=True)
ds.isnull().sum()
x = ds.iloc[:,:-1]
y = ds.iloc[:,-1]
```

```python

x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

lo = LogisticRegression()
lo.fit(x_train,y_train)
lo.score(x_test,y_test)*100
```

```python
from sklearn.metrics import confusion_matrix,f1_score,recall_score,precision_score
clf = confusion_matrix(y_test,lo.predict(x_test))
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(clf,annot=True)
```

```python
precision_score(y_test,lo.predict(x_test))*100
```

```python
f1_score(y_test,lo.predict(x_test))*100
```

```python
recall_score(y_test,lo.predict(x_test))*100
```
