## INHERITANCE
PASSING SOME PROPERTIES FROM ONE CLASS TO ANOTHER CLASS.

```python
class Car:

    def start(self):
        print('car is starting...')

    def stop(self):
        print('car is stopped')

    
class car_name(Car):
    def __init__(self, name):
        self.name = name

bmw = car_name('bmw')

bmw.stop()
print(bmw.name)

```

## MULTI - LEVEL INHERITANCE

```python
class Car:

    def start(self):
        print('car is starting...')

    def stop(self):
        print('car is stopped')

    
class car_name(Car):
    def __init__(self, name):
        self.name = name


class bmw(car_name):
    def __init__(self, model):
        self.model = model


car1 = bmw("veyron")

car1.start()
print(car1.model)
```

## MULTIPLE INHERITANCE
**ALSO The Use Of SUPER( ) In Methods**

```python
class father():

    def __init__(self, phy):
        self.phy = phy
        

    def height(self):
        print('tall')

class mother():
    
    def eye_c(self):
        print('blue eyes')


class child(father , mother):

    def __init__(self, phy):
        super().__init__(phy)

child1 = child('jacked')

print(child1.phy)

```

 ## CLASS METHOD 
 USING THE OBJECT IN DEF TO CHANGE THE OBJECT IN CLASS.

```python
class student():

    name =  'anonymous'

    def new_name(self, name):
        self.name = name

        #self.__class__.name = 'niggr'          # CHANGES THE NAME PRESENT IN THE CLASS.


    # YOU CAN ALSO CHANGE THE NAME PRESENT IN THE CLASS.
    @classmethod
    def change_name(self, name):
        self.name = name

st1 = student()
st1.new_name('nigger')
st1.change_name('niggr')


print(st1.name)

print(student.name)

```

### **PROPERTY METHOD**

```python
class student():

    def __init__(self, phy, chem, maths):
        self.phy = phy
        self.chem = chem
        self.maths = maths

    @property
    def perc(self):
        print('total percenatge : ', (self.phy + self.chem + self.maths)/3 , '%')



st1 = student(97, 98, 99)
print(st1.perc)


st1.phy = 86
print(st1.perc)
```

## **POLYMORPHISM**
In programming, it allows different classes to have methods with the same name, but each class carries out that method in its own specific way.

```python
class complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def show_no(self):
        print(self.real , 'i ' ,'+' , self.img , 'j')


    def __add__(a , b):                 #  __add__ here is a DUNDER Funtion OR another CONSTRUCTOR here
        new_r = a.real + b.real
        new_i = a.img + b.img
        print(f'({new_r} , {new_i})')
        
    

a = complex(3, 4)
# a.show_no()


b =  complex( 7 , 6)

c =  a + b
print(c)
```

## PRACTISE QUESTION


```python
class circle:

    def __init__(self, r):
        self.r = r

    def area(self):
        print(f'Area of the circle of radius {self.r} is {self.r * 3.14 * self.r}')

    def per(self):
        print(f'Perimeter of the circle of radius {self.r} is {self.r * 3.14 * 2}')


c1 = circle(1)

c1.area()
c1.per()
```
