```python
import pandas as pd 

ds = pd.read_csv(r"C:\Users\systems\Downloads\train.zip")
test_data = pd.read_csv(r"C:\Users\systems\Downloads\test.zip")

ds = ds.sample(1000, random_state=42)
ds.shape

ds['pickup_datetime'] = pd.to_datetime(ds['pickup_datetime'])
ds['dropoff_datetime'] = pd.to_datetime(ds['dropoff_datetime'])

# Extract datetime features
ds['pickup_hour'] = ds['pickup_datetime'].dt.hour
ds['pickup_weekday'] = ds['pickup_datetime'].dt.weekday
ds['pickup_month'] = ds['pickup_datetime'].dt.month

# Optional: Calculate trip duration manually and compare
ds['duration_calc'] = (ds['dropoff_datetime'] - ds['pickup_datetime']).dt.total_seconds() / 60


ds.drop(['pickup_datetime', 'dropoff_datetime', 'store_and_fwd_flag'], axis=1, inplace=True)
# Show result
ds.head()

x = ds.iloc[:,1:]

y = ds['trip_duration']

x = ds.drop(['trip_duration' , 'id'], axis = 1)


# dtr or rfr

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)


rc = RandomForestRegressor(random_state=42, criterion='absolute_error', max_depth=4, n_estimators=100)
rc.fit(x_train, y_train)
print(round(rc.score(x_test,y_test)*100,2) , round(rc.score(x_train,y_train)*100,2))


prd = rc.predict(test_data)

prd = pd.DataFrame({'id': test_data['id'], 'trip_duration': prd})

prd

```
