# KNN CLASSIFICATION


```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

ds = pd.read_csv('subs.csv')

x = ds.iloc[:,:-1]
y = ds['purchased']


from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

from sklearn.preprocessing import StandardScaler
std = StandardScaler()
std.fit(x)

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train, y_train)
print(knn.score(x_test, y_test) * 100, knn.score(x_train, y_train) * 100)


for i in range(1,15):
    knn1 = KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train,y_train)

    a = knn1.score(x_train,y_train)*100
    b = knn1.score(x_test,y_test)*100

    print(f'{i}  {a-b}')
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=knn)
```

# KNN REGRESSION

```python
from sklearn.neighbors import KNeighborsRegressor

knn2 = KNeighborsRegressor(9)
knn2.fit(x_train,y_train)


for i in range(1,15):
    knn1 = KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train,y_train)

    a = knn1.score(x_train,y_train)*100
    b = knn1.score(x_test,y_test)*100

    print(f'{i}  {a-b}')
```

```python
print(knn2.score(x_test,y_test)*100)


plot_decision_regions(x.to_numpy(),y.to_numpy(), clf = knn2)
```

# SUPPORT VECTOR MACHINES

```python
from sklearn.svm import SVC

for i in ('linear', 'poly', 'rbf', 'sigmoid'):
    svc = SVC(kernel=i)
    svc.fit(x_train,y_train)

    print(svc.score(x_train,y_train)*100 , svc.score(x_test,y_test)*100)

svc = SVC(kernel='rbf')
svc.fit(x_train,y_train)



plt.figure(figsize=(3,2))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=svc)
```

```python
from sklearn.svm import SVR

for i in ('linear', 'poly', 'rbf', 'sigmoid'):
    svr = SVR(kernel=i)
    svr.fit(x_train,y_train)

    print(svr.score(x_train,y_train)*100 , svr.score(x_test,y_test)*100)

svr = SVR(kernel='rbf')
svr.fit(x_train,y_train)



plt.figure(figsize=(3,2))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=svr)
```
